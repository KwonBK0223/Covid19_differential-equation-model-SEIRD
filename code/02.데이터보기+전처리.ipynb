```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('font', family = 'Gulim')
mpl.rcParams['axes.unicode_minus']=False
```

# 확진자, 사망자수

```python
data1 = pd.read_csv("확진자_사망자.csv", encoding="euc-kr")
```

```python
data1 = data1.iloc[1:]
```

```python
data1["누적확진"] = 0
data1["누적사망"] = 0
```

```python
data1 = data1.replace("-", 0)
```

```python
data1["국내발생(명)"] = data1["국내발생(명)"].astype(int)
data1["해외유입(명)"] = data1["해외유입(명)"].astype(int)
data1["사망(명)"] = data1["사망(명)"].astype(int)
```

```python
data1.info()
```

```python
data1["누적확진"] = data1["계(명)"].cumsum()
data1["누적사망"] = data1["사망(명)"].cumsum()
```

```python
data1
```

```python
data1["계(명)"].plot()
data1["사망(명)"].plot()
```

```python
data1["누적확진"].plot()
plt.legend(loc="upper left")
plt.show()
```

```python
data1["누적사망"].plot()
plt.legend(loc="upper left")
plt.show()
```

```python
data1["누적확진"].plot()
data1["누적사망"].plot()
plt.legend(loc="upper left")
plt.show()
```

# 연령별 확진자 수

```python
data2 = pd.read_csv("연령별_확진자.csv", encoding="euc-kr")
```

```python
data2 = data2.iloc[1:]
```

```python
data2 = data2.replace("-" ,0)
```

```python
data2 = data2.fillna(0)
```

```python
data2.iloc[:,2:] = data2.iloc[:,2:].astype(int)
```

```python
data2.info()
```

```python
data2["0"] = 0
data2["10"] = 0
data2["20"] = 0
data2["30"] = 0
data2["40"] = 0
data2["50"] = 0
data2["60"] = 0
data2["70"] = 0
data2["80"] = 0
```

```python
data2["0"] = data2["0-9세"].cumsum()
data2["10"] = data2["10-19세"].cumsum()
data2["20"] = data2["20-29세"].cumsum()
data2["30"] = data2["30-39세"].cumsum()
data2["40"] = data2["40-49세"].cumsum()
data2["50"] = data2["50-59세"].cumsum()
data2["60"] = data2["60-69세"].cumsum()
data2["70"] = data2["70-79세"].cumsum()
data2["80"] = data2["80세이상"].cumsum()
```

```python
data2
```

```python
data2["0"].plot()
data2["10"].plot()
data2["20"].plot()
data2["30"].plot()
data2["40"].plot()
data2["50"].plot()
data2["60"].plot()
data2["70"].plot()
data2["80"].plot()
plt.legend(loc="upper left")
plt.show()
```

```python
time = time.iloc[1:]
```

```python
time["일자"] = pd.to_datetime(time["일자"])
```

```python
time
```

```python
time["year"] = time["일자"].dt.year
time["month"] = time["일자"].dt.month
time["day"] = time["일자"].dt.day
```

```python
time = time.iloc[:,1:]
```
